Write a program that: Takes a number from the user Checks whether the number is positive, negative, or zero Displays an appropriate message

```python
num=int(input("Enter a number: "))
if num==0:
  print("It is zero.")
elif num<0:
  print("It is negative.")
else:
  print("It is positive number")
```

Write a function that:
Accepts a list of numbers
Returns the sum and average of the list

```python
def reading(num_list):
  sum=0
  length=len(num_list)
  for i in num_list:
    sum=sum+i
  avg=sum/length
  print(f"The sum is {sum}")
  print(f"The average is {avg}")


num_list=[]
num=int(input("Enter the number of elements: "))
for i in range(num):
  number=int(input("Enter the value: "))
  num_list.append(number)
reading(num_list)
```

Write a program using a for loop to print the frst n even numbers where n is taken as input from the user

```python
start=int(input("Enter the start range for elements: "))
st=int(input("Enter the stop range for elements:"))
for i in range(start,st+1):
  if i%2==0:
    print(f"{i} is even")
  else:
    continue
```

Write a program to check whether a given string is a palindrome without using any built-in reverse
function.

```python
value=input("Enter the string: ")
length=len(value)
is_palindrome = True
for i in range(length // 2):
  if value[i] != value[length-i-1]:
    is_palindrome = False
    break

if is_palindrome:
  print("It is a palindrome")
else:
  print("It is not a palindrome")
```

Create a NumPy array with at least 5 integer elements and:
Print the array
Print the maximum and minimum value
Print the sum of all elements

```python
import numpy as np
array=np.array([1,2,3,4,5])
print(array)
print(f"The maximum value is {array.max()}")
print(f"The minimum value is {array.min()}")
print(f"The sum of all the elements is {array.sum()}")
```
