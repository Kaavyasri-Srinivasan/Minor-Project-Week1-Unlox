Project Option 2: Student Marks Analyzer (Recommended)
5.1 Problem Statement
Develop a program that analyzes student marks using Python and NumPy.

```python
import numpy as np
num_of_students=int(input("Enter the number of students: "))
num_of_subjects=int(input("Enter the number of marks: "))
stud_name=np.array([])#empty array using numpy is created to store the respective values.
marks_arr=np.array([])#empty array using numpy is created to store the respective values. This will store total marks for each student.
grade_arr=np.array([])#empty array using numpy is created to store the respective values. This will store grades for each student.

for i in range(num_of_students):
  name=input("Enter the name: ")
  stud_name=np.append(stud_name,name)  #This is to store the values dynamically using numpy in array

  current_student_sum = 0 # Initialize sum for the current student

  for j in range(num_of_subjects):
    while True:
      mark=int(input(f"Enter mark for subject {j+1} for {name}: ")) # More descriptive prompt
      if mark>=0 and mark<=100:  #the marks validation is done here
        current_student_sum += mark  # Accumulate marks for the current student
        break
      else:
        print("Invalid mark.. Mark must be between 0 and 100.")

  marks_arr=np.append(marks_arr, current_student_sum) # Store total sum for the current student

  # Calculate grade based on the average mark for the current student
  average_mark = current_student_sum / num_of_subjects

  if average_mark > 85:
    grade_arr=np.append(grade_arr,'A')
  elif average_mark >= 70 and average_mark <= 85:
    grade_arr=np.append(grade_arr,'B')
  elif average_mark >= 50 and average_mark <= 69:
    grade_arr=np.append(grade_arr,'C')
  else:
    grade_arr=np.append(grade_arr,'Fail')

for i in range(num_of_students):
  print(f"The student {stud_name[i]} secured a total of {marks_arr[i]} with an average of {marks_arr[i]/num_of_subjects:.2f} and a grade of {grade_arr[i]}.")
```
