Write a program that:
Takes a number from the user
Uses a loop to calculate the sum of its digits
Displays the result

```python
num=int(input("Enter the number: ")) #read the input from the user
sum=0
string=str(num)#change the int to str to split each value
for i in string:
  sum=sum+int(i) #the string is changed to int for adding
print(f"The sum of the digit is {sum}")
```

Write a function that:
Accepts a list of integers
Returns a new list containing only even numbers

```python
num=int(input("Enter the number of elements in the list: "))
num_list=[] #empty list to store the elements when read
even_list=[] # empty list to sotre the even numbers
for i in range(num):
  print("Enter the number: ")
  num_list.append(int(input()))  # read the elements in loop from the user
for i in num_list:
  if i%2==0: #check the even condition from the user
    even_list.append(i)
print("The even numbers are: ") #print the even numbers from the even_list[] list
for i in even_list:
  print(i)
```

Write a program to generate the Fibonacci series up to n terms using a loop, where n is taken as user input

```python
num=int(input("Enter the number of inputs: "))  # to read the number of elements in the resultant series
a=0  #default fibonacci initial value
b=1  #default fibonacci second value
series=[] #empty list to store the fibonacci values one by one
series.append(a)
series.append(b)
for i in range(num-2):  #here the range is num-2 because 0 and 1 is given as series in the beginning itself. so their iterations is reduced

  c=a+b  #logic of fibonacci series to add the previous 2 values and the resultant one is the next value
  a=b
  b=c
  series.append(c)
print(series)
```

Write a program that:
Takes a list of numbers from the user
Uses slicing to reverse the list
Prints both the original and reversed list

```python
n=int(input("Enter the number of elements: ")) #read the number of elements for the list from the user
values=[] #empty list for values
print("Enter the values:")
for i in range(n):
  values.append(int(input())) #read the input values from the user
print(f"The original list is: {values}")
print(f"The reverse list is: {values[::-1]}") #slicing operator[::-1] is used to reverse the list
```

Write a program to check whether a given number is a palindrome using arithmetic operations (do
not convert the number to a string).

```python
n=int(input("Enter the number: "))
num=n
rev_num=0 #to store the reversed number

# Reverse the number using arithmetic operations
while num!=0:
  digit=num%10  # to detach the last value from the digit
  rev_num=rev_num*10+digit  #this is used to reverse the number.. this is the logic
  num=num//10 #to detach the values that is other than the mod value we have selected with digit=num%10

# Compare the original number with the reversed number
if n==rev_num:
  print("It is a palindrome.")
else:
  print("It is not a palindrome.")
```
